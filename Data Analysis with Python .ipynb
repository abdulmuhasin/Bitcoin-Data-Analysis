```python
#Use a for loop to iterate over the rows of the DataFrame and calculate the daily price
#change (difference between `Close` and `Open`), adding this as a new column `Price
#Change
price_chge = []
```

```python
for index, row in df.iterrows():
    price_change = row['Close'] - row['Open']
    price_chge.append(price_change)
```

```python
df['Price Change'] = price_chge
```

```python
df
```

```python
#Apply an if-else condition to create another column `Price Trend`, where:
#If the `Close` price is higher than the `Open` price, set the value to "Up."
#Otherwise, set the value to "Down."
price_trnd = []
```

```python
for index, row in df.iterrows():
    if row['Close'] > row['Open']:
        price_trend = "Up"
    else:
        price_trend = "Down"
    price_trnd.append(price_trend)
```

```python
df['Price Trend'] = price_trnd
```
